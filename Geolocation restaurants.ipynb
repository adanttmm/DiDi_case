# Analize geographical restaurant locations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/adanttmm/Documents/DiDi test/Files&graphs/geoloc_rests.csv')
```

```python
BBox = ((130.000,145.000,      
         32.000, 45.000))
```

```python
bub_size = 200
max_visitors = max(df.avg_visitors)

ruh_m = plt.imread('/home/adanttmm/Documents/DiDi test/Files&graphs/map.png')
fig, ax = plt.subplots(figsize = (8,7))
scatter = ax.scatter(df.longitude, df.latitude,c='r',s=df.avg_visitors*bub_size/max_visitors, alpha = 0.2)
ax.set_title('Restaurants geolocation')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
# produce a legend with the unique colors from the scatter
kw = dict(prop="sizes", num=5, color='r', fmt="{x:.2f}",
          func=lambda s: np.sqrt(s/.3)/3)
legend2 = ax.legend(*scatter.legend_elements(**kw),
                    loc="lower right", title="Visits")
```
